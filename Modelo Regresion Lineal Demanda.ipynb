```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Step 1: Generate the Dataset
np.random.seed(0)  # for reproducibility
sales_data = np.random.randint(50, 100, size=30)  # simulate 30 days of sales data
print("Sales data for the last 30 days:")
print(sales_data)

```

```python
# Step 2: Build the Model
# Create a pandas Series
sales_series = pd.Series(sales_data)

# Fit the ARIMA model
model = ARIMA(sales_series, order=(5,1,0))  # order=(p,d,q)
model_fit = model.fit()

```

```python
# Step 3: Make Predictions
forecast = model_fit.forecast(steps=5)
print("\nPredicted sales for the next 5 days:")
print(forecast)

```

```python
# Optional: Plot the results
plt.figure(figsize=(10, 6))
plt.plot(sales_series, label='Sales Data (Last 30 days)')
plt.plot(np.arange(30, 35), forecast, label='Predicted Sales (Next 5 days)', color='red')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Sales Forecast for the Next 5 Days')
plt.show()
```
